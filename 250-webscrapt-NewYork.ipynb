```python
import urllib3                     # 避免认证警告
urllib3.disable_warnings()

import requests

from bs4 import BeautifulSoup

r=requests.get("https://century21.com/real-estate/new-york-city-ny/LCNYNEWYORKCITY/?", verify=False) 
c=r.content

soup=BeautifulSoup(c,"html.parser")


all=soup.find_all("div",{"class":"property-card-primary-info"}) 
```

```python
l=[]
base_url="https://century21.com/real-estate/new-york-city-ny/LCNYNEWYORKCITY/?p="
for page in range(0,3,1):
      print(base_url+str(page))                            #可以生成对应的多个网页
      r=requests.get(base_url+str(page),verify=False)       #将 verify=False 加到此处，避免error
      c=r.content
      soup=BeautifulSoup(c,"html.parser")
      all=soup.find_all("div",{"class":"property-card-primary-info"}) 
      for item in all:
            d={}
            try:
                    d["Address"]=item.find("div",{"class":"property-address"}).text.replace("\n","").replace(" ","")
            except:
                    print("")

            try:
                    d["Price"]=item.find("a",{"class":"listing-price"}).text.replace("\n","").replace(" ","")
            except:
                    print("")

            try:       
                    d["Beds"]=item.find("div",{"class":"property-beds"}).find("strong").text
            except:
                    print("")

            try: 
                    d["Baths"]=item.find("div",{"class":"property-baths"}).find("strong").text
            except:
                    print("")       
                    
            try:        
                    d["Half-baths"]=item.find("div",{"class":"property-half-baths"}).find("strong").text
            except:
                    print("")

            l.append(d)         #将创建的dictionary 加入到library中
```

```python
import pandas                           #通过pandas创建数据表
df=pandas.DataFrame(l)
```
